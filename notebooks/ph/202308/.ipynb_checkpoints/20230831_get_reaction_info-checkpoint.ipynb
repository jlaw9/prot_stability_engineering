### Try to figure out for each enzyme which direction is the main reaction
- Some enzymes may have been tested in the reverse direction of the reaction listed in BRENDA.
- Here we will try to figure out which direction is the "main" direction for each enzyme and assume that is what was tested experimentally
- For a given organism, if there are no products listed then assume the reaction goes in the forward direction

#### What if there are multiple substrates and products
- Take the sum (mean?) of the predictions and use whichever is largest

```python
def map_ligands_to_inchi(ligands):
    inchis = [ligand_to_inchi.get(l) for l in ligands]
    return inchis
    # if None in inchis:
        # return np.nan
```

```python
def get_subst_inchi(df):
    df = df.copy()
    df['subst_list'] = df.substrates.apply(lambda x: [s.replace('2 ', '') for s in x.split(' + ')])
    df['subst_inchi'] = (df.subst_list
                     .apply(map_ligands_to_inchi)
                     .apply(lambda x: np.nan if None in x else x))
    return df
```

```python
def get_subst_prod_no_cofactors(df, cofactors):
    """ Get only rows where both substrates and products map to inchi strings.
    This will be used to try and figure out which reaction is the "main" reaction
    assuming the reaction is reversible
    """
    df = df.copy()
    df = df.dropna(subset='products')
    df['subst_list'] = df.substrates.apply(lambda x: [s.replace('2 ', '') for s in x.split(' + ')])
    df['subst_list_no_cof'] = df.subst_list.apply(lambda x: set(x) - cofactors)
    idx_cofactors_only = df[df.subst_list.apply(lambda x: len(set(x) - set(cofactors))) == 0].index
    df.loc[idx_cofactors_only, 'subst_list_no_cof'] = df.loc[idx_cofactors_only, 'subst_list']
    
    # df['prod_list'] = df.products.apply(lambda x: [s.replace('2 ', '') for s in x.split(' + ')] 
    #                                                     if isinstance(x, str) else x)    
    # df['prod_list_no_cof'] = df.prod_list.apply(lambda x: set(x) - cofactors
    #                                            if isinstance(x, list) else x)
    # idx_cofactors_only = df[df.prod_list.apply(lambda x: len(set(x) - set(cofactors))
    #                                            if isinstance(x, list) else -1) == 0].index
    # df.loc[idx_cofactors_only, 'prod_list_no_cof'] = df.loc[idx_cofactors_only, 'prod_list']
    df['prod_list'] = df.products.apply(lambda x: [s.replace('2 ', '') for s in x.split(' + ')])
    df['prod_list_no_cof'] = df.prod_list.apply(lambda x: set(x) - cofactors)
    idx_cofactors_only = df[df.prod_list.apply(lambda x: len(set(x) - set(cofactors))) == 0].index
    df.loc[idx_cofactors_only, 'prod_list_no_cof'] = df.loc[idx_cofactors_only, 'prod_list']
    
    # now map to inchi
    df['subst_inchi'] = (df.subst_list_no_cof
                         .apply(map_ligands_to_inchi)
                         .apply(lambda x: np.nan if None in x else x))
    df['prod_inchi'] = (df.prod_list_no_cof
                        .apply(map_ligands_to_inchi)
                        .apply(lambda x: np.nan if None in x else x))

    return df
```

  - c. If the organism is not listed for the reaction, predict if the substrate or product is the main substrate for each enzyme
- 2. proteins with multiple exp_index
  - Hopefully the comments will be useful to differentiate

#### Combine the "natural substrates" with the "substrates" tables
- For some reason the "natural substrates" table doesn't have the reference, so if the substrate table has the reference, then use that reaction

```python
def get_enzyme_reactions(uniprot_ec, data_ph, data_rxn, data_nat_rxn):
    single_substrate = []
    multi_substrate = []
    no_substrate = []
    for i, (ec_num, u_id) in enumerate(tqdm(uniprot_ec.index)):
        row = data_ph[(data_ph.ec_num == ec_num) & 
                      (data_ph.uniprot_id == u_id)]

        rxns = data_rxn[data_rxn.ref.isin([int(r) for r in row.ref.values[0].replace(' ','').split(',')])]
        substrates = set(list(rxns.substrates.values))
        products = set(list(rxns.products.values))
        if len(substrates) == 1:
            row['substrates'] = rxns.substrates.values[0]
            row['products'] = rxns.products.values[0]
            single_substrate += [row]
            continue    
        # if there isn't a single reaction for this paper, 
        # then check if there is a single "natural substrate" reaction listed for this organism. 
        # If not, then track the rest of the possible reactions

        # get only the substrates that map to a molecular structure
        nat_rxns = data_nat_rxn[(data_nat_rxn.ec_num == ec_num) &
                                (data_nat_rxn.organism == row.organism.values[0])]
        if len(rxns) == 1:
            row['substrates'] = nat_rxns.products.values[0]
            row['products'] = nat_rxns.products.values[0]
            single_substrate += [row]
            continue

        if len(nat_rxns) > 1:
            substrates |= set(list(nat_rxns.substrates.values))
            products |= set(list(nat_rxns.products.values))
        # if there's no substrates listed for this organism, then use the substrates from the other organisms
        if len(substrates) == 0:
            rxns = data_nat_rxn[(data_nat_rxn.ec_num == ec_num)]
            substrates = set(list(rxns.substrates.values))
            products = set(list(rxns.products.values))
        if len(substrates) == 0:
            rxns = data_rxn[(data_rxn.ec_num == ec_num)]
            substrates = set(list(rxns.substrates.values))
            products = set(list(rxns.products.values))

        if len(substrates) == 1:
            row['substrates'] = list(substrates)[0]
            row['products'] = list(products)[0]
            # data_ph_rxn = pd.concat([data_ph_rxn, row])
            single_substrate += [row]
        elif len(substrates) > 1:
            row['substrates'] = [substrates]
            # the reaction may be reversible, so keep track of the products as well
            row['products'] = [products]
            multi_substrate += [row]
        else: 
            no_substrate += [row]

    single_substrate = pd.concat(single_substrate)
    multi_substrate = pd.concat(multi_substrate)
    no_substrate = pd.concat(no_substrate)
    print(f"{len(single_substrate) = }, {len(multi_substrate) = }, {len(no_substrate) = }")
    return single_substrate, multi_substrate, no_substrate

single_substrate, multi_substrate, no_substrate = get_enzyme_reactions(uniprot_ec_singles, 
                                                                       data_ph_range, 
                                                                       data_rxn_subst_inchi, 
                                                                       data_nat_rxn_subst_inchi)
```

```python
def get_enzyme_reactions_multi(data, data_rxn, data_nat_rxn):
    """ Match the enzyme & ec number with the reaction.
    This function is for cases where there may be multiple reactions tested for a single protein
    """
    # single_substrate = []
    data_prot_ec_multi_rxn = []
    data_prot_ec_multi_rxn_no_substrate = []
    for (ec_num, u_id, refs), rows in data.groupby(['ec_num', 'uniprot_id', 'ref']):

        rxns = data_rxn[data_rxn.ref.isin([int(r) for r in refs.replace(' ','').split(',')])]
        substrates = set(list(rxns.substrates.values))
        products = set(list(rxns.products.values))
        if len(substrates) == 1:
            rows['substrates'] = [substrates]*len(rows)
            rows['products'] = [products]*len(rows)
            data_prot_ec_multi_rxn += [rows]
            continue

        nat_rxns = data_nat_rxn[(data_nat_rxn.ec_num == ec_num) &
                                (data_nat_rxn.organism == rows.organism.values[0])]

        if len(nat_rxns) >= 1:
            substrates |= set(list(nat_rxns.substrates.values))
            products |= set(list(nat_rxns.products.values))
        # if there's no substrates listed for this organism, then use the substrates from the other organisms
        if len(substrates) == 0:
            nat_rxns = data_nat_rxn_subst_inchi[(data_nat_rxn_subst_inchi.ec_num == ec_num)]
            substrates = set(list(nat_rxns.substrates.values))
            products = set(list(nat_rxns.products.values))
        # print(nat_rxns)

        if len(substrates) >= 1:
            rows['substrates'] = [substrates]*len(rows)
            # the reaction may be reversible, so keep track of the products as well
            rows['products'] = [products]*len(rows)
            data_prot_ec_multi_rxn += [rows]
        else: 
            data_prot_ec_multi_rxn_no_substrate += [rows]
        # if i > 100:
            # break

    data_prot_ec_multi_rxn = pd.concat(data_prot_ec_multi_rxn)
    data_prot_ec_multi_rxn_no_substrate = pd.concat(data_prot_ec_multi_rxn_no_substrate)
    print(f"{len(data_prot_ec_multi_rxn) = }, {len(data_prot_ec_multi_rxn_no_substrate) = }")
    return data_prot_ec_multi_rxn, data_prot_ec_multi_rxn_no_substrate
    
data_prot_ec_multi_rxn, data_prot_ec_multi_rxn_no_substrate = get_enzyme_reactions_multi(data_uniprot_ec_multi, 
                                                                                         data_rxn_subst_inchi, 
                                                                                         data_nat_rxn_subst_inchi)
```

Now map the substrates and products to inchi (again)

```python
def sort_and_unique_substrates(substrate_sets):
    sort_uniq_subst = []
    for substrates in substrate_sets:
        substr_list = substrates.split(' + ')
        # sometimes a number of the molecule is included (e.g., 'inulin + 5 H2O')
        for i in range(2, 7):
            substr_list = [s.replace(f'{i} ', '') for s in substr_list]
        substr_list = sorted(substr_list)
        substr_list = ' + '.join(substr_list)
        sort_uniq_subst += [substr_list]
    sort_uniq_subst = [x for x in sort_uniq_subst if '?' not in x]
    sort_uniq_subst = set(sort_uniq_subst)
    sort_uniq_subst = [s.split(' + ') for s in sort_uniq_subst]
    return sort_uniq_subst
```

```python
def combine_and_map_to_inchi(df):
    df['subst_list'] = df.substrates.apply(sort_and_unique_substrates)
    df['prod_list'] = df.products.apply(sort_and_unique_substrates)
    # # now map to inchi
    # df['subst_inchi'] = (df.subst_list_no_cof
    #                      .apply(map_ligands_to_inchi)
    #                      .apply(lambda x: np.nan if None in x else x))
    # df['prod_inchi'] = (df.prod_list_no_cof
    #                     .apply(map_ligands_to_inchi)
    #                     .apply(lambda x: np.nan if None in x else x))
    
    def combine_subst_prod(row):
        # keep track of the mapping so we can easily go back to the substrate name
        combined = {}
        for substrates in row.subst_list:
            for s in substrates:
                inchi = ligand_to_inchi.get(s)
                if s is not None:
                    combined[inchi] = s
                # else:
                    # print(
                
        for products in row.prod_list:
            for s in products:
                inchi = ligand_to_inchi.get(s)
                if s is not None:
                    combined[inchi] = s
        return combined
    
    df['all_subst_prod_map'] = df.apply(combine_subst_prod, axis=1)
    # df['all_subst_prod'] = df['all_subst_prod_map'].apply(lambda x: d
    return df
```

### Oxidoreductases
- many enzymes can perform both the forward and reverse reaction
- From the substrate and products, figure out if this is an oxidation or reduction reaciton

```python
# TODO try using rdkit.Chem.rdMolDescriptors.CalcOxidationNumbers
# https://www.rdkit.org/docs/source/rdkit.Chem.rdMolDescriptors.html
```

```python
oxidoreductases = data_subst[data_subst.ec_num.apply(lambda x: x.split('.')[0] == "1")]
print(len(oxidoreductases))
oxidoreductases.head(2)
```

```python
def get_substrate_set(substrate_sets):
    substrate_set = set()
    if isinstance(substrate_sets, str):
        substrate_sets = [substrate_sets]
    for substrates in substrate_sets:
        substr_list = substrates.split(' + ')
        # sometimes a number of the molecule is included (e.g., 'inulin + 5 H2O')
        for i in range(2, 7):
            substr_list = [s.replace(f'{i} ', '') for s in substr_list]
        substrate_set |= set(substr_list)
    return substrate_set
        
oxidoreductases['all_substrates'] = oxidoreductases.substrates.apply(get_substrate_set)
oxidoreductases.head(2)
```

```python
oxidation_cofactors = ['NAD(P)+', 
                       'NAD+',
                       'NADP+',
                       'FAD',
                       'H2O',
                       'O2',
                       'acceptor',
                      ]

reduction_cofactors = ['NAD(P)H', 
                       'NADH',
                       'NADPH',
                       'FAD2H',
                       'H2O2',
                       'reduced acceptor',
                      ]
```

```python
# convert the substrate sets to strings for now so we can just search for specific cofactors
# doesn't work for subsets e.g., FAD & FADH2 or H20 & H202
# oxidoreductases['subst_str'] = oxidoreductases.substrates.astype(str)
def guess_rxn_type(subst_set):
    oxidation = False
    reduction = False
    for oc in oxidation_cofactors:
        if oc in subst_set:
            oxidation = True
    for rc in reduction_cofactors:
        if rc in subst_set:
            reduction = True
    
    if oxidation and reduction:
        return "oxidation_or_reduction"
    elif oxidation:
        return "oxidation"
    elif reduction:
        return "reduction"
    else:
        return "not_found"
    
oxidoreductases['rxn_type_guess'] = oxidoreductases.all_substrates.apply(guess_rxn_type)
print(oxidoreductases.rxn_type_guess.value_counts())
```
